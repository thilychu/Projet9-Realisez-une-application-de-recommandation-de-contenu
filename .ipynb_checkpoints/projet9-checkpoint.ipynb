## Content-Based Filtering 
Le Content-Based Filtering (Filtrage basÃ© sur le contenu en franÃ§ais) est une mÃ©thode de recommandation utilisÃ©e dans les systÃ¨mes de recommandation pour proposer des Ã©lÃ©ments similaires Ã  un utilisateur en se basant sur les caractÃ©ristiques ou le contenu des Ã©lÃ©ments qu'il a dÃ©jÃ  apprÃ©ciÃ©s. 

    Extraction des caractÃ©ristiques des Ã©lÃ©ments : Tout d'abord, il est essentiel d'extraire des caractÃ©ristiques pertinentes des Ã©lÃ©ments Ã  recommander. 

    Construction du profil de l'utilisateur : En suite on construit un profil utilisateur basÃ© sur les interactions comme favoris, aimÃ©s ou consultÃ©s. 

    Calcul de la similaritÃ© : On calcule la similaritÃ© cosinus entre le profil de l'utilisateur et les caractÃ©ristiques des Ã©lÃ©ments disponibles. 

    Recommandations personnalisÃ©es : Les Ã©lÃ©ments qui prÃ©sentent la plus grande similaritÃ© avec le profil de l'utilisateur sont alors recommandÃ©s. Plus la similaritÃ© est Ã©levÃ©e, plus l'Ã©lÃ©ment a de chances d'Ãªtre recommandÃ©.

**ModÃ©lisation des prÃ©fÃ©rences de l'utilisateur**

Le rating de l'utilisateur pour un article est dÃ©fini comme la moyenne de tous les articles cliquÃ©s:

    Rating(user,article)= #Clicks(user,article) / #Clicks(user)


```python
def aggregate_articles(articles):
    return articles.groupby(lambda x: True).agg(
        {
            col: "mean"
            if is_numeric_dtype(articles.dtypes[col])
            else lambda x: x.mode()[0]
            for col in articles.columns
        }
    )


def get_user_interest(user_id, clicks, articles):
    all_article_ids = clicks.query("user_id == @user_id")["click_article_id"]
    interest = aggregate_articles(articles.query("article_id in @all_article_ids")).drop(["article_id"], axis=1)
    return interest
```

**Calculer la similaritÃ© cosinus (cosine_similarity) entre le profil utilisateur et les articles**

Les caractÃ©ristiques avec des Ã©chelles diffÃ©rentes peuvent avoir un impact disproportionnÃ© sur la mesure de similaritÃ©. En normalisant les donnÃ©es, on se assure que toutes les caractÃ©ristiques ont la mÃªme importance dans le calcul de la similaritÃ©

```python
def normalize_data(interest, articles):
    articles_copy = articles.copy()
    articles_copy["created_at_ts"] = articles["created_at_ts"].apply(lambda x: x.value)
    articles_copy = articles_copy.drop(["article_id"], axis=1, errors="ignore")
    interest_copy = interest.copy()
    interest_copy["created_at_ts"] = interest_copy["created_at_ts"].apply(lambda x: x.value)
    # CrÃ©ez le scaler et normalisez le DataFrame
    scaler = StandardScaler()
    articles_std = scaler.fit_transform(articles_copy)
    interest_std = scaler.transform(interest_copy)
    
    return interest_std, articles_std
```

```python
def get_closest_articles(user_id, clicks, articles, n=10):
    
    interest = get_user_interest(user_id, clicks, articles)
    interest_std, articles_std = normalize_data(interest, articles)
    articles["similarity"] = cosine_similarity(interest_std, articles_std)[0]
    closest_articles = articles.sort_values("similarity", ascending=False).iloc[:n]
    
    return closest_articles
```

## Collaborative Filtering

Il existe deux principales approches de filtrage collaboratif :

1. Filtrage collaboratif basÃ© sur les utilisateurs :

Dans cette approche, les recommandations sont gÃ©nÃ©rÃ©es en identifiant des utilisateurs similaires Ã  l'utilisateur cible. La similaritÃ© peut Ãªtre calculÃ©e en fonction des notations ou des interactions passÃ©es avec les Ã©lÃ©ments.
Une fois les utilisateurs similaires identifiÃ©s, les Ã©lÃ©ments qui ont Ã©tÃ© apprÃ©ciÃ©s par ces utilisateurs et qui n'ont pas encore Ã©tÃ© apprÃ©ciÃ©s par l'utilisateur cible sont recommandÃ©s Ã  ce dernier.
Par exemple, si l'utilisateur A et l'utilisateur B ont des prÃ©fÃ©rences similaires pour les films et que l'utilisateur A a aimÃ© un film que l'utilisateur B n'a pas encore vu, le systÃ¨me recommandera ce film Ã  l'utilisateur B.

2. Filtrage collaboratif basÃ© sur les Ã©lÃ©ments :

Dans cette approche, les recommandations sont gÃ©nÃ©rÃ©es en se basant sur les caractÃ©ristiques des Ã©lÃ©ments et en trouvant des Ã©lÃ©ments similaires Ã  ceux que l'utilisateur a dÃ©jÃ  aimÃ©s.
Les caractÃ©ristiques des Ã©lÃ©ments peuvent inclure des genres, des mots-clÃ©s, des catÃ©gories, etc. Par exemple, pour les films, les caractÃ©ristiques peuvent inclure le genre, le rÃ©alisateur, les acteurs, etc.
Lorsqu'un utilisateur a aimÃ© un Ã©lÃ©ment, le systÃ¨me recommandera d'autres Ã©lÃ©ments similaires en fonction de leurs caractÃ©ristiques.
Par exemple, si un utilisateur a aimÃ© un film de science-fiction, le systÃ¨me recommandera d'autres films de science-fiction.
Le filtrage collaboratif peut Ãªtre mis en Å“uvre de diffÃ©rentes maniÃ¨res, notamment avec des mÃ©thodes de calcul de similaritÃ©, telles que la similaritÃ© cosinus ou la distance euclidienne. De plus, il peut Ãªtre utilisÃ© avec des donnÃ©es explicites (notations ou Ã©valuations numÃ©riques donnÃ©es par les utilisateurs) ou des donnÃ©es implicites (comme les clics, les vues, les achats, etc.).
